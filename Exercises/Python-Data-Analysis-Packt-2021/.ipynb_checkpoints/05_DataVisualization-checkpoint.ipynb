```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df=pd.read_csv("Datasets/HR_comma_sep.csv")
```

```python
df.head()
```

```python
# Create lmplot
sns.lmplot(x='satisfaction_level', y='last_evaluation', data=df, fit_reg=False, hue='left')
# Display plot
plt.show()
```

```python
# Violin plot
sns.violinplot(data=df[['satisfaction_level','last_evaluation']])
```

```python
# Count plot
sns.countplot(x='salary', data=df, hue='left')
```

```python
# Joint plot using kde
sns.jointplot(x='satisfaction_level', y='last_evaluation', data=df, kind='kde')
```

```python
# Heatmap
# Load data
data = sns.load_dataset("iris")
# Correlation matrix
corr_matrix = data.corr()
# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
```

```python
# Pair plot the iris data
sns.pairplot(data)
```

```python
# Import the required modules
from bokeh.plotting import figure
from bokeh.plotting import output_notebook, show
from bokeh.layouts import row, column
# Import iris flower dataset as pandas DataFrame
from bokeh.sampledata.iris import flowers as df
# Output to notebook
output_notebook()
# Instantiate a figure
fig1 = figure(plot_width = 300, plot_height = 300)
fig2 = figure(plot_width = 300, plot_height = 300)
fig3 = figure(plot_width = 300, plot_height = 300)
# Create scatter marker plot by render the circles
fig1.circle(df['petal_length'], df['sepal_length'], size=8, color ="green", alpha = 0.5)
fig2.circle(df['petal_length'], df['sepal_width'], size=8, color = "blue", alpha = 0.5)
fig3.circle(df['petal_length'], df['petal_width'], size=8, color = "red", alpha = 0.5)
# Create row layout
row_layout = row(fig1, fig2, fig3)
# Show the plot
show(row_layout)
```

```python
# Interactions - Hide Click Policy
# Import the module
from bokeh.plotting import figure
from bokeh.plotting import output_notebook
from bokeh.plotting import show
from bokeh.models import CategoricalColorMapper

# Import sample dataset
from bokeh.sampledata.iris import flowers as df

# Output to notebook
output_notebook()

# Create a figure object
fig = figure(plot_width=500,
             plot_height=350,
             title='Petal Length vs. Petal Width',
             x_axis_label='petal_length',
             y_axis_label='petal_width'
            )

# Create scatterplot
for specie, color in zip(['setosa', 'virginica', 'versicolor'],['blue', 'green', 'red']):
    data=df[df.species==specie]
    fig.circle('petal_length',
               'petal_width',
               size=8,
               color=color,
               alpha=0.7,
               legend_label=specie,
               source=data
              )
# Hide click policy
fig.legend.location='top_left'
fig.legend.click_policy='hide'

# Show graph
show(fig)
```

```python
# Interactions - Muted Click Policy
# Import the module
from bokeh.plotting import figure
from bokeh.plotting import output_notebook
from bokeh.plotting import show
from bokeh.models import CategoricalColorMapper

# Import sample dataset
from bokeh.sampledata.iris import flowers as df

# Output to notebook
output_notebook()

# Create a figure object
fig = figure(plot_width=500,
             plot_height=350,
             title='Petal Length vs. Petal Width',
             x_axis_label='petal_length',
             y_axis_label='petal_width'
            )

# Create scatterplot
for specie, color in zip(['setosa', 'virginica', 'versicolor'],['blue', 'green', 'red']):
    data=df[df.species==specie]
    fig.circle('petal_length',
               'petal_width',
               size=8,
               color=color,
               alpha=0.7,
               legend_label=specie,
               source=data,
               muted_color=color,
               muted_alpha=0.2
              )
# Hide click policy
fig.legend.location='top_left'
fig.legend.click_policy='mute'

# Show graph
show(fig)
```

```python
# Annotations
# Import modules
from bokeh.plotting import figure
from bokeh.plotting import output_notebook
from bokeh.plotting import show
from bokeh.models import CategoricalColorMapper

# Import sample dataset
from bokeh.sampledata.iris import flowers as df

# Output to notebook
output_notebook()

# Create a color mapper for categorical columns
color_mapper=CategoricalColorMapper(factors=['setosa','virginica','versicolor'],
                                    palette=['blue','green','red']
                                   )
color_dict={'field':'species',
            'transform':color_mapper
           }

# Create a figure object
p=figure(plot_width=500,
         plot_height=350,
         title='Petal Length vs. Petal Width',
         x_axis_label='petal_length',
         y_axis_label='petal_width'
        )

# Create a scatterplot
p.circle('petal_length',
         'petal_width',
         size=8,
         color=color_dict,
         alpha=0.5,
         legend_group='species',
         source=df
        )

# Set the legend location
p.legend.location='top_left'

# Show graph
show(p)
```

```python
# Hover tool
# Import modules
from bokeh.plotting import figure
from bokeh.plotting import output_notebook
from bokeh.plotting import show
from bokeh.models import CategoricalColorMapper
from bokeh.models import HoverTool

# Import sample dataset
from bokeh.sampledata.iris import flowers as df

# Output to notebook
output_notebook()

# Create a color mapper for categorical columns
color_mapper=CategoricalColorMapper(factors=['setosa','virginica','versicolor'],
                                    palette=['blue','green','red']
                                   )
color_dict={'field':'species',
            'transform':color_mapper
           }

# Create a hover tool and specify hovering information
hover=HoverTool(tooltips=[('Species type','@species'),
                          ('IRIS Petal Length','@petal_length'),
                          ('IRIS Petal Width','@petal_width')
                         ])

# Create a figure object
p=figure(plot_width=500,
         plot_height=350,
         title='Petal Length vs. Petal Width',
         x_axis_label='petal_length',
         y_axis_label='petal_width',
         tools=[hover, 'pan', 'wheel_zoom']
        )

# Create a scatterplot
p.circle('petal_length',
         'petal_width',
         size=8,
         color=color_dict,
         alpha=0.5,
         legend_group='species',
         source=df
        )

# Set the legend location
p.legend.location='top_left'

# Show graph
show(p)
```

```python
# Widgets - Tab Panel
# Import modules
from bokeh.plotting import figure
from bokeh.plotting import output_notebook
from bokeh.plotting import show
from bokeh.models.widgets import Tabs
from bokeh.models.widgets import Panel

# Import iris flower dataset
from bokeh.sampledata.iris import flowers as df

# Output to notebook
output_notebook()

# Create a figure
fig1 = figure(plot_width=300, plot_height=300)
fig2 = figure(plot_width=300, plot_height=300)

# Create scatterplot
fig1.circle(df['petal_length'],
            df['sepal_length'],
            size=8,
            alpha=0.5,
            color='green'
           )
fig2.circle(df['petal_length'],
            df['sepal_width'],
            size=8,
            alpha=0.5,
            color='blue'
           )

# Create panels
tab1 = Panel(child=fig1, title='tab1')
tab2 = Panel(child=fig2, title='tab2')

# Create tab
tab_layout = Tabs(tabs=[tab1,tab2])

# Show the plot
show(tab_layout)
```

```python
# Widgets - Slider
# Import modules
from bokeh.plotting import figure
from bokeh.plotting import output_notebook
from bokeh.plotting import show
from bokeh.models import CustomJS
from bokeh.models import ColumnDataSource
from bokeh.models import Slider
from bokeh.layouts import column

# Show output
output_notebook()

# Create a list of data
x = [x for x in range(0,100)]
y = x

# Create a dataframe
df = ColumnDataSource(data={'x_values':x, 'y_values':y})

# Create a figure object
fig = figure(plot_width=350, plot_height=350)

# Create a line plot
fig.line('x_values',
         'y_values',
         source=df,
         line_width=2.5,
         line_alpha=0.8
        )

# Create a callback using CustomJS
callback = CustomJS(args=dict(source=df), code="""
    var data = source.data;
    var f = cb_obj.value
    var x_values = data['x_values']
    var y_values = data['y_values']
    for (var i = 0; i < x_values.length; i++) {
        y_values[i] = Math.pow(x_values[i], f)
    }
    source.change.emit();
""")

slider_widget = Slider(start=0.0,
                       end=10,
                       value=1,
                       step=.1,
                       title='Display the Power of x'
                      )
slider_widget.js_on_change('value', callback)

# Create layout
slider_widget_layout = column(fig, slider_widget)

# Display the layout
show(slider_widget_layout)
```
